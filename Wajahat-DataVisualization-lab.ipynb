# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'ConvertedComp' column from the 'master' table
QUERY = """
SELECT ConvertedComp
FROM master
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot the histogram of 'ConvertedComp'
plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
plt.hist(df['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')  # Plot the histogram with 30 bins
plt.xlabel('Converted Compensation')  # Set the label for the x-axis
plt.ylabel('Frequency')  # Set the label for the y-axis
plt.title('Histogram of Converted Compensation')  # Set the title of the plot
plt.grid(True)  # Show grid lines
plt.show()  # Display the plot
```

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'Age' column from the 'master' table
QUERY = """
SELECT Age
FROM master
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot the box plot of 'Age'
plt.figure(figsize=(8, 6))  # Set the size of the figure (width, height)
plt.boxplot(df['Age'], vert=False)  # Plot the box plot horizontally
plt.xlabel('Age')  # Set the label for the x-axis
plt.title('Box Plot of Age')  # Set the title of the plot
plt.grid(True)  # Show grid lines
plt.show()  # Display the plot
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'Age' and 'WorkWeekHrs' columns from the 'master' table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot the scatter plot of 'Age' versus 'WorkWeekHrs'
plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
plt.scatter(df['Age'], df['WorkWeekHrs'], color='skyblue', alpha=0.6)  # Plot the scatter plot
plt.xlabel('Age')  # Set the label for the x-axis
plt.ylabel('WorkWeekHrs')  # Set the label for the y-axis
plt.title('Scatter Plot of Age vs. WorkWeekHrs')  # Set the title of the plot
plt.grid(True)  # Show grid lines
plt.show()  # Display the plot
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns from the 'master' table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Create the bubble plot
plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)

# Scatter plot with 'WorkWeekHrs' on the x-axis, 'CodeRevHrs' on the y-axis, and 'Age' as bubble size
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.5, color='skyblue')

plt.xlabel('WorkWeekHrs')  # Set the label for the x-axis
plt.ylabel('CodeRevHrs')  # Set the label for the y-axis
plt.title('Bubble Plot: WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')  # Set the title of the plot

# Add a colorbar legend to represent the bubble sizes (Age)
cbar = plt.colorbar()
cbar.set_label('Age', rotation=270, labelpad=15)

plt.grid(True)  # Show grid lines
plt.show()  # Display the plot
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the count of each database that respondents wish to learn next year
QUERY = """
SELECT DatabaseWorkedWith, COUNT(*) as Count
FROM master
WHERE DatabaseWorkedWith IS NOT NULL
GROUP BY DatabaseWorkedWith
ORDER BY Count DESC
LIMIT 5
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Data for the pie chart
databases = df['m4_survey_data.sqlite']
counts = df['Count']
total_count = counts.sum()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=databases, autopct=lambda pct: f"{pct:.1f}% ({pct*total_count/100:.0f})", startangle=140, colors=plt.cm.Paired.colors)
plt.axis('equal')
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the median WorkWeekHrs and CodeRevHrs for the age group 30 to 39
QUERY = """
SELECT Age,
       AVG(CASE WHEN Age >= 30 AND Age <= 39 THEN WorkWeekHrs END) AS Median_WorkWeekHrs,
       AVG(CASE WHEN Age >= 30 AND Age <= 39 THEN CodeRevHrs END) AS Median_CodeRevHrs
FROM master
WHERE Age >= 30 AND Age <= 39
GROUP BY Age
ORDER BY Age
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Data for the stacked chart
ages = df['Age']
median_workweekhrs = df['Median_WorkWeekHrs']
median_coderevhrs = df['Median_CodeRevHrs']

# Plot the stacked chart
plt.figure(figsize=(10, 6))
plt.bar(ages, median_workweekhrs, label='Median WorkWeekHrs', color='orange')
plt.bar(ages, median_coderevhrs, bottom=median_workweekhrs, label='Median CodeRevHrs', color='skyblue')
plt.xlabel('Age Group (30 to 39)')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 39')
plt.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the median ConvertedComp for ages from 45 to 60
QUERY = """
SELECT Age,
       AVG(CASE WHEN Age >= 45 AND Age <= 60 THEN ConvertedComp END) AS Median_ConvertedComp
FROM master
WHERE Age >= 45 AND Age <= 60
GROUP BY Age
ORDER BY Age
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Data for the plot
ages = df['Age']
median_convertedcomp = df['Median_ConvertedComp']

# Plot the median ConvertedComp for all ages from 45 to 60
plt.figure(figsize=(10, 6))
plt.plot(ages, median_convertedcomp, marker='o', linestyle='-', color='purple')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.grid(True)

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to retrieve the count of each category in the column MainBranch
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
WHERE MainBranch IS NOT NULL
GROUP BY MainBranch
ORDER BY Count DESC
"""

# Execute the query and store the result in a Pandas DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Data for the horizontal bar chart
main_branches = df['MainBranch']
counts = df['Count']

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(main_branches, counts, color='skyblue')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.title('Number of Respondents in Each MainBranch Category')
plt.gca().invert_yaxis()  # Invert y-axis to display categories from top to bottom

plt.show()
```
