Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with the 'ConvertedComp' column
# For example:
# df = pd.read_csv('your_dataset.csv')

# Replace 'ConvertedComp' with the actual name of the column you want to calculate the IQR for
column_name = 'ConvertedComp'

# Calculate the first quartile (Q1) of the 'ConvertedComp' column
q1 = df[column_name].quantile(0.25)

# Calculate the third quartile (Q3) of the 'ConvertedComp' column
q3 = df[column_name].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

print("Interquartile Range (IQR) for the column '{}': {}".format(column_name, iqr))
```

Find out the upper and lower bounds.


```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with the 'ConvertedComp' column
# For example:
# df = pd.read_csv('your_dataset.csv')

# Replace 'ConvertedComp' with the actual name of the column you want to calculate bounds for
column_name = 'ConvertedComp'

# Calculate the first quartile (Q1) of the 'ConvertedComp' column
q1 = df[column_name].quantile(0.25)

# Calculate the third quartile (Q3) of the 'ConvertedComp' column
q3 = df[column_name].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Calculate the lower bound
lower_bound = q1 - 1.5 * iqr

# Calculate the upper bound
upper_bound = q3 + 1.5 * iqr

print("Lower Bound for the column '{}': {}".format(column_name, lower_bound))
print("Upper Bound for the column '{}': {}".format(column_name, upper_bound))
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
import pandas as pd

column_name = 'ConvertedComp'

# Calculate the first quartile (Q1) of the 'ConvertedComp' column
q1 = df[column_name].quantile(0.25)

# Calculate the third quartile (Q3) of the 'ConvertedComp' column
q3 = df[column_name].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Calculate the lower bound
lower_bound = q1 - 1.5 * iqr

# Calculate the upper bound
upper_bound = q3 + 1.5 * iqr

# Find the outliers in the 'ConvertedComp' column
outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

# Count the number of outliers
num_outliers = len(outliers)

print("Number of outliers in the column '{}': {}".format(column_name, num_outliers))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
import pandas as pd

column_name = 'ConvertedComp'

# Calculate the first quartile (Q1) of the 'ConvertedComp' column
q1 = df[column_name].quantile(0.25)

# Calculate the third quartile (Q3) of the 'ConvertedComp' column
q3 = df[column_name].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Calculate the lower bound
lower_bound = q1 - 1.5 * iqr

# Calculate the upper bound
upper_bound = q3 + 1.5 * iqr

# Create a new DataFrame without outliers
df_no_outliers = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]

# Print the shape of the new DataFrame to see the number of rows after removing outliers
print("Shape of the new DataFrame without outliers: {}".format(df_no_outliers.shape))
```
